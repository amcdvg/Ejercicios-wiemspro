```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_angle(a, b, c):
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    if angle > 180.0:
        angle = 360 - angle
    return angle
```

```python
A = np.array([1, 3])
B = np.array([3, 1])
C = np.array([5, 4])


angle_B = calculate_angle(A, B, C)

plt.figure(figsize=(5, 5))
plt.plot([A[0], B[0]], [A[1], B[1]], 'bo-', label="AB")
plt.plot([B[0], C[0]], [B[1], C[1]], 'ro-', label="BC")
plt.plot([A[0], C[0]], [A[1], C[1]], 'go-', label="AC")
plt.text(A[0], A[1], ' A', fontsize=12, verticalalignment='bottom', horizontalalignment='right')
plt.text(B[0], B[1], ' B', fontsize=12, verticalalignment='top', horizontalalignment='right')
plt.text(C[0], C[1], ' C', fontsize=12, verticalalignment='bottom', horizontalalignment='left')


plt.text(B[0] + 0.2, B[1] + 0.2, f'{angle_B:.2f}°', fontsize=12, color='purple')

plt.xlim(0, 6)
plt.ylim(0, 5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Triangle and Calculated Angle at B")
plt.legend()
plt.grid(True)
plt.show()
```
